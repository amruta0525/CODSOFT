```python
import random
from tkinter import *

class RockPaperScissors:
    def __init__(self):
        self.player_choice = None
        self.computer_choice = None
        self.user_score = 0
        self.computer_score = 0

    def gui(self):
        root = Tk()
        root.title("Rock-Paper-Scissors")

        frame = Frame(root)
        frame.pack()

        rock_button = Button(frame, text="Rock", command=lambda: self.play_game('rock'))
        rock_button.pack(side=LEFT)

        paper_button = Button(frame, text="Paper", command=lambda: self.play_game('paper'))
        paper_button.pack(side=LEFT)

        scissor_button = Button(frame, text="Scissor", command=lambda: self.play_game('scissors'))
        scissor_button.pack(side=LEFT)

        play_again_button = Button(root, text="Play Again", command=self.play_again)
        play_again_button.pack()

        self.score_label = Label(root, text="Score: User - 0 | Computer - 0")
        self.score_label.pack()

        root.mainloop()

    def play_game(self, choice):
        choices = ['rock', 'paper', 'scissors']
        self.player_choice = choice
        self.computer_choice = random.choice(choices)
        result = self.winner_determination(self.player_choice, self.computer_choice)

        if result == "You Lose!":
            self.computer_score += 1
        elif result == "You Win!":
            self.user_score += 1

        self.update_score_label()

    def winner_determination(self, player_choice, computer_choice):
        if player_choice == computer_choice:
            return "It's a tie!"
        elif player_choice == 'rock':
            return "You Win!" if computer_choice == 'scissors' else "You Lose!"
        elif player_choice == 'paper':
            return "You Win!" if computer_choice == 'rock' else "You Lose!"
        elif player_choice == 'scissors':
            return "You Win!" if computer_choice == 'paper' else "You Lose!"

    def update_score_label(self):
        self.score_label.config(text=f"Score: User - {self.user_score} | Computer - {self.computer_score}")

    def play_again(self):
        self.user_score = 0
        self.computer_score = 0
        self.update_score_label()
        print("New Game", "Scores reset. Start a new game!")

if __name__ == "__main__":
    game = RockPaperScissors()
    game.gui()
```
